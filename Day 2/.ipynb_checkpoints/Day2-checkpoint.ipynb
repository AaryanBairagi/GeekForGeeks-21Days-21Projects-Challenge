```python
##STEP 2 : CREATE A DATAFRAME FOR THE DATASET
netflix_df = pd.read_csv("netflix_titles.csv")
```

```python
#checking the basic info regarding the dataset
netflix_df.head()
```

```python
##OBSERVATION : GIVEN DATASET HAS MISSING VALUES IN - 'director' , 'cast', 'country', 'date_added', 'rating' attributes
```

```python
##STEP 3 : DATA CLEANING AND TRANSFORMATION 
netflix_df.isnull()
```

```python
netflix_df.isnull().sum()
```

```python
#1. since the director and cast are string so we can add "Unknown" placeholder there instead of dropping these rows
netflix_df['director'] = netflix_df['director'].fillna('Unknown')
netflix_df['cast'] = netflix_df['cast'].fillna('Unknown')
```

```python
#2. to handle missing values in the Country column, we can use the mode (most frequent value), assuming that the mode represents the most probable value for the missing entries
mode_country = netflix_df['country'].mode()[0] #since its in datafrajme format
netflix_df['country'] = netflix_df['country'].fillna(mode_country) 
```

```python
#3. to handle rating and date_added we can now drop the null rows because honestly its a large dataset so dropping them wont affect the dataset
netflix_df.dropna(subset=['date_added','rating'], inplace = True)
```

```python
#check any more missing values are remaining? 
netflix_df.isnull().sum()
```

```python
##the date_added attribute has values of type Object so we change there value to date-time
netflix_df['date_added'] = pd.to_datetime(netflix_df['date_added'] , format = 'mixed' , dayfirst = False)
```

```python
netflix_df['date_added']
```

```python
##year and month are important sources of information to us so extract it from these dates
netflix_df['year_added'] = netflix_df['date_added'].dt.year
netflix_df['month_added'] = netflix_df['date_added'].dt.month

netflix_df.head()
```

```python
##STEP 4 : IMPLEMENTING EDA(Exploratory Data Analysis)
#1. Pie Chart showing the proportions of the Movie and TV Show
plt.figure(figsize=(8,6))
type_counts = netflix_df['type'].value_counts()
plt.pie(type_counts,labels=type_counts.index,autopct='%1.1f%%',startangle=140,colors=['cyan','orange'])
plt.title('Proportions of Movies and TV Shows :')
plt.show()
```

```python
#OBSERVATION : Netflix library has 70% movies and 30% TV shows
```

```python
#2.Analysing how the content was added over time 
content_added_as_per_year = netflix_df.groupby(['year_added','type']).size().unstack().fillna(0)
content_added_as_per_year
```

```python
#plotting a line plot for this data
plt.figure(figsize=(14,8))
content_added_as_per_year.plot(kind='line',marker='o',figsize=(14,8))
plt.title("Content added to Netflix over the years")
plt.xlabel('Year Added')
plt.ylabel('Number of Titles Added')
plt.legend(title='Content-Type')
plt.grid(True)
plt.show()
```

```python
#OBSERVATION : There has been a steady growth in the number of contents added on Netflix over the years, peaking around 2018–2019. However, due to the impact of COVID-19, the content additions slightly declined afterward.
```

```python
#3. listed_in is a csv attribute we need rows corresponding to each seperate genre
genres = netflix_df.assign(genre = netflix_df['listed_in'].str.split(', ')).explode('genre')
genres
```

```python
##the above dataframe is entirely stored in genres variable
top_genres = genres['genre'].value_counts().reset_index()
top_genres.columns = ['genre','count']
```

```python
top_15_genres = top_genres.head(15)
top_15_genres
```

```python
#plot a bar plot for this data 
plt.figure(figsize=(18,12))
sns.barplot(x='count',y='genre',data=top_15_genres,palette='mako',hue='genre',legend=False)
plt.title('Top 15 Genres on Netflix')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
#OBSERVATION : Interntaional Movies is the top genre in Netflix library
```

```python
#4. distribution of duration in dataset is either in seasons or minutes/hours for TV shows and Movies respectively so we anayze them seperately
movie_df = netflix_df[netflix_df['type']=='Movie'].copy()
tv_show_df = netflix_df[netflix_df['type']=='TV Show'].copy()
```

```python
#clean duration for movie_df
movie_df['duration_min'] = movie_df['duration'].str.replace(' min','').astype(int)  ##converting '93 min ' to '93'

#clean duration for tv_show_df
tv_show_df['seasons'] = tv_show_df['duration'].str.replace(' Seasons','').str.replace(' Season','').astype(int) ##converting '4 Seasons' and '1 Season' to '4' and '1'
```

```python
movie_df.head()
```

```python
tv_show_df.head()
```

```python
#plot the distributions
fig , axes = plt.subplots(1,2,figsize=(18,7))

#Movies Duration Distribution
sns.histplot(ax=axes[0],x='duration_min',data=movie_df,kde=True,color='skyblue',bins=50).set_title('Movies Duration Distribution(mins)')

#TV Show Duration Distribution
sns.countplot(ax=axes[1],x='seasons',data=tv_show_df,palette='rocket',hue='seasons',order=tv_show_df['seasons'].value_counts().index,legend=False).set_title('TV Show Duration Distribution(seasons)')
plt.show()
```

```python
#Observation: The majority of movies have a runtime between 90–120 minutes, while most TV shows consist of only 1 season.
```

```python
#5.Geograhical Analysis : majority if the records in the dataset have multiple country values in csv format
countries = netflix_df.assign(country = netflix_df['country'].str.split(', ')).explode('country')
countries
```

```python
top_country = countries['country'].value_counts().reset_index()
top_country.columns = ['country','count']
top_country
```

```python
top_15_country = top_country.head(15)
top_15_country
```

```python
#plot the distribution on a barplot
plt.figure(figsize=(12,10))
sns.barplot(data=top_15_country,x='count',y='country',palette='mako',hue='country',legend=False)
plt.title('Top 15 Content Producing Countries on Netflix :')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

```python
#OBSERVATION : USA is the major content generating country for Netflix
```

```python
#6. Maturity Rating Distribution
plt.figure(figsize=(12,8))
sns.countplot(x='rating',data=netflix_df,palette='crest',hue='rating',legend=False, order=netflix_df['rating'].value_counts().index)
plt.title('Distribution of Content Ratings on Netflix')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
#OBSERVATION : A large amount of content on Netflix is rated TV-MA 
```

```python
##STEP 5 : FETAURE ENGINEERING - Content Freshness
netflix_df['age'] = netflix_df['year_added'] - netflix_df['release_year']
netflix_df['age'].head()
```

```python
content_age = netflix_df[netflix_df['age'] >= 0]
content_age
```

```python
#STEP 6 : DEEPER MULTIVARIATE ANALYSIS
##finding all the outliers in the top 5 genres

genre_frame = genres['genre'].value_counts().index[:5]
genre_frame
```

```python
top_5_genre_frame['duration_min'] = top_5_genre_frame['duration'].str.replace(' min','').astype(int)
```

```python
#OBSERVATION : The Median for top Gneres is quite similar : 90min-100min
```
