```python
##STEP 6 : DATA PREPROCESSING AND FEATURE ENGINEERING
#Combine both train and test data for consistent preprocessing

all_data = pd.concat((train_df.loc[:,:'SaleCondition'],test_df.loc[:,:'SaleCondition']))
print('Combined Data Dimensions:',all_data.shape)
```

```python
all_data.head()
```

```python
##STEP 7 : HANDLING MISSING VALUES
all_data.isnull().sum()
```

```python
missing_value = all_data.isnull().sum().sort_values(ascending=False)
missing_value
```

```python
missing_value = missing_value[missing_value>0]
missing_value
```

```python
#1. Fill the numeric attributes with 0 
numerical_cols_to_impute_zero = ['MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'GarageCars', 'GarageArea', 'BsmtFullBath', 'BsmtHalfBath']
#if these numeric columns ar present in all_data then fil the null values that respective columns with 0 
for col in numerical_cols_to_impute_zero:
    if col in all_data.columns:
        all_data[col] = all_data[col].fillna(0)
```

```python
all_data[numerical_cols_to_impute_zero].isnull().sum()
```

```python
#2. Impute the LotFrontage with the median of the neighbourhood
##filling both the columns that have been grouped by with their median values
if 'LotFrontage' in all_data.columns and all_data['LotFrontage'].isnull().any():
    all_data['LotFrontage'] = all_data.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))
```

```python
all_data['Alley'].value_counts()
```

```python
# 3. Impute the categorical variables with 'None' (whenre NA means no data) or mode (where NA means missing data)
categorical_cols_to_impute_none = ['Alley', 'Fence', 'MiscFeature', 'PoolQC', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'MasVnrType']
for col in categorical_cols_to_impute_none:
    if col in all_data.columns:
        all_data[col] = all_data[col].fillna('None')

categorical_cols_to_impute_mode = ['Electrical', 'KitchenQual', 'Exterior1st', 'Exterior2nd', 'SaleType', 'Utilities', 'Functional', 'MSZoning']
for col in categorical_cols_to_impute_mode: 
    if col in all_data.columns:
        all_data[col] = all_data[col].fillna(all_data[col].mode()[0])
```

```python
# 4. Impute the GarageYrBlt with 0 (0 means no garage , consistent with GarageArea)
if 'GarageYrBlt' in all_data.columns:
    all_data['GarageYrBlt'] = all_data['GarageYrBlt'].fillna(0)
```

```python
all_data.isnull().sum()
```

```python
#total sum of all these rows can also be done
all_data.isnull().sum().sum()
```

```python
##STEP 8 : FEATURE ENGINEERING
#total squre foot area
all_data['TotalSqrFT'] = all_data['TotalBsmtSF'] + all_data['1stFlrSF'] + all_data['2ndFlrSF']

#total age of the house
all_data['Age'] = all_data['YrSold'] - all_data['YearBuilt']

#total bathroom area
all_data['BathFeature'] = (all_data['FullBath'] + (0.5 * all_data['HalfBath']) + all_data['BsmtFullBath'] + (0.5 * all_data['BsmtHalfBath']))
```

```python
##STEP 9 : CATEGORICAL ENCODING
#convert all the object attributes into a binary - True or False (suitable for model training and testing) ,i.e, implement One-hot Encoding
categorical_cols = all_data.select_dtypes(include=['object']).columns
all_data = pd.get_dummies(all_data,columns=categorical_cols,drop_first=True)
##drop_first means drop one category at a time to avoid multi-collinearity
```

```python
all_data.shape
```

```python
all_data.head()
```

```python
##STEP 10 : MODEL TRAINING AND TESTING
#earlier we combined both the train_df and test_df for consistent preprocessing now we split them again into training and testing dataset
X = all_data[:len(train_df)]  ## all the entries from 0 to last entry in training df
Y = train_df['SalePrice']
X_test_final = all_data[len(train_df):] ##all the entries after training df last entry

x_train , x_test , y_train , y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
##STEP 11 : FEATURE SCALING
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

X_test_final_scaled = scaler.transform(X_test_final)
```

```python
##STEP 12 : LINEAR REGRESSION
lr = LinearRegression()
lr.fit(x_train_scaled,y_train)

y_predicted = lr.predict(x_test_scaled)
```

```python
#plot the linear regression 
plt.figure(figsize=(12,8))
plt.scatter(y_test,y_predicted,alpha=0.6,color='skyblue')
plt.title('Actual VS Predicted Sale Price')
plt.xlabel('Actual Sale Price')
plt.ylabel('Predicted Sale Price')
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()],'r--') #prediction line first array is X-coordinates and second array is Y-coordinates and 'r' means red line '--' means dashed line
plt.show()
```

```python
#imputing COST Functions
mse = mean_squared_error(y_test,y_predicted)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test,y_predicted)
r2 = r2_score(y_test,y_predicted)

print(f"MSE:",mse)
print(f"Root MSE:",rmse)
print(f"MAE:",mae)
print(f"R2 SCORE:",r2)
```

```python
# STEP 13 : XGBOOST MODEL (advanced)
xgbr = xgb.XGBRegressor(
    objective='reg:squarederror', #decide what task to perform
    n_estimators=1000, #number of trees to build
    learning_rate=0.05, #contribution of each tree
    max_depth=3, #depth of each tree
    min_child_weight=1, #minimum sum of weights in a child node
    subsample=0.8, #fraction of sample used for each tree
    colsample_bytree=0.8,  #fraction of features used per tree
    random_state=42 
)
```

```python
xgbr.fit(x_train,y_train)
y_pred_xgbr = xgbr.predict(x_test)
```

```python
#imputing COST Functions
mse = mean_squared_error(y_test,y_pred_xgbr)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test,y_pred_xgbr)
r2 = r2_score(y_test,y_pred_xgbr)

print(f"MSE:",mse)
print(f"Root MSE:",rmse)
print(f"MAE:",mae)
print(f"R2 SCORE:",r2)
```

```python
# Make predictions on the final test set using the best model (XGBoost)
# Use the unscaled test data for XGBoost prediction
final_predictions_log = xgbr.predict(X_test_final)

# IMPORTANT: We must reverse the log transformation to get the predictions back on the original price scale
final_predictions = np.expm1(final_predictions_log)

# Create the submission DataFrame
submission = pd.DataFrame({'Id': test_df.index, 'SalePrice': final_predictions})

# Save to csv
submission.to_csv('submission.csv', index=False)

print("Submission file 'submission.csv' created successfully.")
submission.head()
```
