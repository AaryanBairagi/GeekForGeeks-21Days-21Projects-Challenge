# **STEP 1 : Load Model and Tokenizer**

```python
##IMPORTING ALL THE LIBRARIES
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer , pipeline
```

```python
model_name = 'gpt2'
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
generator = pipeline('text-generation' , model=model , tokenizer=tokenizer , device=0 if torch.cuda.is_available() else -1)
```

# **STEP 2 : Filtering Mechanism**

```python
def check_python_query(prompt:str) -> bool :
  """
  Very simple keyword-based filter to check if the prompt
  is about Python coding.
  """
  keywords = ["python", "code", "function", "class", "def", "import", "list", "dict"]
  prompt_lower = prompt.lower()
  return any(kw in prompt_lower for kw in keywords)
```

# **STEP 3 : Handle User Query**

```python
def handle_user_query(prompt:str , max_length: int=100) :
  if (check_python_query(prompt)):
    response = generator(prompt,max_length=max_length,num_return_sequences=1,do_sample=True)
    print(response)
    return response[0]['generated_text']
  else :
    return "This model can only answer python related queries."
```

# **STEP 4 : TESTING**

```python
##TESTING ON PYTHON QUERY
prompt1 = "Write a Python function to reverse a string."
print(prompt1)
response  = handle_user_query(prompt1)
print(response)
```

```python
##TESTING ON NORMAL QUERY
prompt2 = "Whats the capital of India?"
print(prompt2)
response2 = handle_user_query(prompt2)
print(response2)
```
